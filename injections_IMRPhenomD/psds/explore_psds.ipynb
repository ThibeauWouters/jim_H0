```python
import numpy as np
import matplotlib.pyplot as plt 

import xml.etree.ElementTree as ET
```

```python
def fetch_numbers_from_stream(element):
    for child in element.iter('Stream'):
        # Get the text content of the <Stream> element
        stream_text = child.text.strip()
        # Split the text content by whitespace to get individual numbers
        values = stream_text.split()
        numbers = [float(values[i]) for i in range(1, len(values), 2)]

        return np.array(numbers)

# Find the <Stream> element and fetch the numbers
for filename in ["./O4HL_psd.xml", "./O5HL_psd.xml"]:
    tree = ET.parse(filename)
    root = tree.getroot()
    numbers = fetch_numbers_from_stream(root)
    
    plt.plot(numbers)
    plt.yscale('log')
    plt.show()
```
