```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
posterior = "./outdir/injection_GW170817/results_production.npz"

NAMING = ['M_c', 'q', 's1_z', 's2_z', 'lambda_1', 'lambda_2', 'H_0', 'z', 't_c', 'phase_c', 'cos_iota', 'psi', 'ra', 'sin_dec']

data = np.load(posterior)
chains = data['chains']
chains = np.reshape(chains, (-1, len(NAMING)))


H0_index = NAMING.index('H_0')
z_index = NAMING.index('z')

H0_posterior = chains[:,H0_index]
z_posterior = chains[:,z_index]

def get_dL(H0, z):
    c = 299792.458 # km/s
    dL = c * z / H0
    return dL
```

```python
dL_posterior = get_dL(H0_posterior, z_posterior)

plt.hist(dL_posterior, bins=100, density=True)
plt.show()
```
